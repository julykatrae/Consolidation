```python
tt_class_3 = titanic["class"] == "3rd"
tt_c3_total = titanic[tt_class_3]["survived"].count() 
tt_class_3_survived = (titanic["class"] == "3rd") & (titanic["survived"] == "yes")
tt_c3_survived = titanic[tt_class_3_survived]["survived"].count() 
tt_c3_survival = tt_c3_survived / tt_c3_total * 100
tt_c3_survival
```

```python
passengers = titanic[titanic["class"].isin(["1st", "2nd", "3rd"])]
```

```python
passengers
```

```python
def survived_ratio(titanic):
    yes = titanic[titanic["survived"] == "yes"]
    return len(yes) / len(titanic)

ratio_1st = survived_ratio(passengers[passengers["class"] == "1st"])
ratio_2nd = survived_ratio(passengers[passengers["class"] == "2nd"])
ratio_3rd = survived_ratio(passengers[passengers["class"] == "3rd"])

print(ratio_1st, ratio_2nd, ratio_3rd)
```

```python
ratio_males = survived_ratio(titanic[titanic["gender"] == "male"])
ratio_females = survived_ratio(titanic[titanic["gender"] == "female"])
print(ratio_males, ratio_females)
```

```python
ratio_children = survived_ratio(titanic[titanic["age"] < 18])
ratio_adults = survived_ratio(titanic[titanic["age"] >= 18])
print(ratio_children, ratio_adults)
```

```python
ratio_parch = survived_ratio(titanic[titanic["parch"] > 0])
ratio_sibsp = survived_ratio(titanic[titanic["sibsp"] > 0])
ratio_alone = survived_ratio(titanic[(titanic["sibsp"] == 0) & (titanic["parch"] == 0)])
print(ratio_parch, ratio_sibsp, ratio_alone)
```
